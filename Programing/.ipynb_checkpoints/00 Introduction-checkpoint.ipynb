```python
from scipy.integrate import odeint, ode
```

```python
help(odeint)
```

```python
def dx(x, t):
    
    " x=[x1,x2,x3,x4]"
    g, L, m = 9.82, 0.5, 0.1
    
    x1, x2, x3, x4 = x
    c1 = 1 / (m * L**2)
    dx1 = 6.0 * c1 * (2 * x3 -3 * cos(x1 - x2) * x4) / (16.0 -9.0 * cos(x1 - x2)**2)
    dx2 = 6.0 * c1 * (8 * x4 -3 * cos(x1 - x2) * x3) / (16.0 -9.0 * cos(x1 - x2)**2)
    dx3 = -0.5 / c1 * (dx1 * dx2 * sin(x1 - x2) + 3 * g / L * sin(x1))
    dx4 = -0.5 / c1 * (-dx1 * dx2 * sin(x1 - x2) + g / L * sin(x2))
    
    return array([dx1, dx2, dx3, dx4])
```

```python
x0 = [pi / 4, pi / 2, 0, 0]
t = linspace(0, 10, 250)
```

```python
x = odeint(dx, x0, t)
```

```python
plt.plot(t, x[:,0], label='theta1')
plt.plot(t, x[:,1], label='theta2')
plt.legend(loc='best')
```

```python
L = 0.5
x1 = L * sin(x[:, 0])
y1 = -L * cos(x[:, 0])
x2 = x1 + L * sin(x[:, 1])
y2 = y1 - L * cos(x[:, 1])

plt.plot(x1, y1, label = 'pendulum1')
plt.plot(x2, y2, label = 'pendulum2')
```

```python
import time
from IPython.display import clear_output

fig, ax = plt.subplots(figsize = (4, 4))

for t_idx, tt in enumerate(t[:200]):

    x1 = + L * sin(x[t_idx, 0])
    y1 = - L * cos(x[t_idx, 0])

    x2 = x1 + L * sin(x[t_idx, 1])
    y2 = y1 - L * cos(x[t_idx, 1])
    
    ax.cla()    
    ax.plot([0, x1], [0, y1], 'r.-')
    ax.plot([x1, x2], [y1, y2], 'b.-')
    ax.set_ylim([-1.5, 0.5])
    ax.set_xlim([1, -1])

    clear_output() 
    display(fig)

    time.sleep(0.1)
```

```python
def dy(y, t, zeta, w0):
    x, p = y
    dx = p
    dp = -2 * zeta * w0 * p - w0**2 * x
    return array([dx, dp])
```

```python
y0 = [1.0, 0.0]
t = linspace(0, 10, 1000)
w0 = 2 * pi
```

```python
y1 = odeint(dy, y0, t, args = (0.0, w0)) # undamped
y2 = odeint(dy, y0, t, args = (0.2, w0)) # under damped
y3 = odeint(dy, y0, t, args = (1.0, w0)) # critical damping
y4 = odeint(dy, y0, t, args = (5.0, w0)) # over damped
```

```python
plt.plot(t, y1[:, 0], label='undamped')
plt.plot(t, y2[:, 0], label='under damped')
plt.plot(t, y3[:, 0], label='critical damping')
plt.plot(t, y4[:, 0], label='over damped')
plt.legend(loc='best')
```
